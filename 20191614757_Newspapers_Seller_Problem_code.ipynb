```python
# get the news type and demand

def RN(nt,d):
    if 35 >= nt >= 1:
        if 3 >= d >= 1:
            return 40;
        elif 8 >= d >= 4 :
            return 50;
        elif 23 >= d >= 9:
            return 60;
        elif 43 >= d >= 24:
            return 70;
        elif 78 >= d >= 44:
            return 80;
        elif 93 >= d >= 79:
            return 90;
        elif 99 >= d >= 94 or d == 0:
            return 100;
    if 80 >= nt >= 36:
        if 10 >= d >= 1:
            return 40;
        elif 28 >= d >= 11 :
            return 50;
        elif 68 >= d >= 29:
            return 60;
        elif 88 >= d >= 69:
            return 70;
        elif 96 >= d >= 89:
            return 80;
        elif 99 >= d >= 97 or d == 0:
            return 90;
    if 99 >= nt >= 81 or nt == 0:
        if 44 >= d >= 1:
            return 40;
        elif 66 >= d >= 45 :
            return 50;
        elif 82 >= d >= 67:
            return 60;
        elif 94 >= d >= 83:
            return 70;
        elif 99 >= d >= 95 or d == 0:
            return 80;
```

```python
#  calculate Revenue from sales

def rev(p,demand):
    if p <= demand :
        return p * 0.5 ;
    else:
        return demand * 0.5;
```

```python
# calcuate lost profit

def lose_prof(p,demand):
    if demand > p :
        return (demand - p) * 0.17;
    else:
        return 0;
```

```python
#calculate scrap revenue

def scrap(p,demand):
    if p > demand:
        return (p - demand) * 0.07;
    else:
        return 0;
```

```python
# get daily profit for purchase item daily

def purch(p,nt,d):
    purchase = p
    cost_of_np = purchase * 0.33
    daily_profit = []
    for i in range(0,len(nt)) :
        demand = RN(nt[i],d[i])
        print("day {}\ndemand = {}".format(i+1,demand))
        rev_sales = rev(purchase,demand)
        print("revenue = {}".format(rev_sales))
        lost_prof = lose_prof(purchase,demand)
        print("lost profit = {}".format(lost_prof))
        scrap_rev = scrap(purchase,demand)
        print("scrap revenue = {}".format(scrap_rev))
        daily_profit.append(rev_sales - cost_of_np - lost_prof + scrap_rev)
        print("day {} profit = {}\n".format(i+1,daily_profit[i]))
    return sum(daily_profit);
```

```python
# get the highest profit

def high_prof(x,y,z):
    if x > y and x > z:
        return x,50;
    elif y > x and y > z:
        return y,60;
    else:
        return z,70;
```
